```python
import pandas as np
import networkx as nx
import matplotlib.pyplot as plt

# load dataset
with open('./brain_edges_.csv') as f: 
    lines = f.readlines();

# lines are edges in our graph
lines = [line.strip().split() for line in lines]
# lines
```

```python
# plot main graph
import matplotlib.pyplot as plt
plt.figure(figsize=(30,20))

# create graph
G = nx.Graph()
G.add_edges_from(lines)

nx.draw_spring(G,
               with_labels=True,
               arrows=False,
               alpha=1,
               edge_color='r',
               node_label_size=10,
               node_color='black',
               node_size=12)
nx.number_of_nodes(G)
```

```python
# barabasi model
barabasi_model = nx.barabasi_albert_graph(n=785, m=15, seed=None)

plt.figure(figsize = (8, 8))
nx.draw_spring(barabasi_model,
           with_labels=True,
           alpha=1,
           edge_color='lightgreen',
           node_label_size=3,
           node_color='b',
           node_size=8)
print(len(barabasi_model.edges()))
```

```python
#compare average clustering
averageClusteringBrain = nx.average_clustering(G)
averageClusteringBarabasi = nx.average_clustering(barabasi_model)


print('average clustering brain: ', round(averageClusteringBrain, 4))
print('average clustering barabasi: ', round(averageClusteringBarabasi, 4))
```

```python
#compare degree distributions
import collections

def plot_degree_distribution(G):
    degree_sequence = sorted([d for n, d in G.degree()], reverse=True)  # degree sequence
    degreeCount = collections.Counter(degree_sequence)
    deg, cnt = zip(*degreeCount.items())
    fig, ax = plt.subplots()
    plt.bar(deg, cnt)
```

```python
# for barin network
plot_degree_distribution(G)
```

```python
# for barabasi network
plot_degree_distribution(barabasi_model)
```

```python
# compare diameters
brainDublinDiameter = nx.diameter(G)
barabasiDublinDiameter = nx.diameter(barabasi_model)

print('brain diameter: ', round(brainDublinDiameter, 4))
print('barabasi diameter: ', round(barabasiDublinDiameter, 4))
```

```python
# compare densities
brainDensity = nx.density(G)
barabasiDensity = nx.density(barabasi_model)

print('brain density: ', round(brainDensity, 4))
print('barabasi density: ', round(barabasiDensity, 4))
```

```python
# compare average shortest path length
brainShortetstPathLength = nx.average_shortest_path_length(G)
barabasiShortetstPathLength = nx.average_shortest_path_length(barabasi_model)

print('brain shortest path length: ', round(brainShortetstPathLength, 4))
print('barabasi shortest path length: ', round(barabasiShortetstPathLength, 4))
```

```python
# compare assortativity coefficient
brianAssortativity = nx.degree_assortativity_coefficient(G)
barabasiAssortativity = nx.degree_assortativity_coefficient(barabasi_model)

print('brain assortativity coefficient: ', round(brianAssortativity, 4))
print('barabasi assortativity coefficient: ', round(barabasiAssortativity, 4))
```

```python
# compare transitivities
brainTransitivity = nx.transitivity(G)
barabasiTransitivity = nx.transitivity(barabasi_model)

print('brain transitivity: ', round(brainTransitivity, 4))
print('barabasi transitivity: ', round(barabasiTransitivity, 4))
```
